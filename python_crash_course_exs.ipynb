# Chapter 9: Classes

```python
#9-1
class Restaurant:
    """A class representing a restaurant."""

    def __init__(self, name, cuisine_type):
      """Initialize the restaurant."""
      self.name = name
      self.cuisine_type = cuisine_type

    def describe_restaurant(self):
      print(f"The name of the restaurant is {self.name}")
      print(f"The cuisine type of the restaurant is {self.cuisine_type}")

    def open_restaurant(self):
      print(f"{self.name} is open")
```

```python
restaurant = Restaurant("Asato", "Sushi")

print(restaurant.name)
print(restaurant.cuisine_type)

restaurant.describe_restaurant()
restaurant.open_restaurant()
```

```python
#9-2
restaurant1 = Restaurant("KFC", "Fast Food")

restaurant1.describe_restaurant()

restaurant2 = Restaurant("McDonalds", "Fast Food")

restaurant2.describe_restaurant()

restaurant3 = Restaurant("Kentucky", "Pizza")

restaurant3.describe_restaurant()
```

```python
#9-3
class User:
  """A class representing a user."""

  def __init__(self, first_name, last_name, phone_number, email, country):
    """Initialize the user."""
    self.first_name = first_name
    self.last_name = last_name
    self.phone_number = phone_number
    self.email = email
    self.country = country

  def describe_user(self):
    """Describe the user"""
    print(f"User summary:\n"
      f"First name: {self.first_name}, Last name: {self.last_name}\n"
      f"Phone number: {self.phone_number}\n"
      f"Email: {self.email}\n"
      f"Country: {self.country}")

  def greet_user(self):
    """Greet the user"""
    print(f"Greetings, {self.first_name} {self.last_name} from {self.country}!")
```

```python
user1 = User("Teo", "Spring", 123, "teo@gmail.com", "Argentina")

user1.describe_user()
user1.greet_user()

user2 = User("Juan", "Perez", 321, "juanperez@outlook.com", "Colombia")

user2.describe_user()
user2.greet_user()

user3 = User("Marta", "Sanchez", 890, "marta_s@gmail.com.ar", "Argentina")

user3.describe_user()
user3.greet_user()
```

```python
#9-4
class Restaurant:
    """A class representing a restaurant."""

    def __init__(self, name, cuisine_type):
      """Initialize the restaurant."""
      self.name = name
      self.cuisine_type = cuisine_type
      self.number_served = 0

    def describe_restaurant(self):
      print(f"The name of the restaurant is {self.name}")
      print(f"The cuisine type of the restaurant is {self.cuisine_type}")
      print(f"The number of customers served is {self.number_served}")

    def open_restaurant(self):
      print(f"{self.name} is open")

    def set_number_served(self, served):
      self.number_served = served

    def increment_number_served(self, customers):
      self.number_served += customers
```

```python
restaurant = Restaurant ("Venice", "Bar")

restaurant.describe_restaurant()

restaurant.number_served = 10

restaurant.describe_restaurant()

restaurant.set_number_served(30)

restaurant.describe_restaurant()

restaurant.increment_number_served(30)

restaurant.describe_restaurant()
```

```python
#9-5
class User:
  """A class representing a user."""

  def __init__(self, first_name, last_name, phone_number, email, country):
    """Initialize the user."""
    self.first_name = first_name
    self.last_name = last_name
    self.phone_number = phone_number
    self.email = email
    self.country = country
    self.login_attempts = 0

  def describe_user(self):
    """Describe the user"""
    print(f"User summary:\n"
      f"First name: {self.first_name}, Last name: {self.last_name}\n"
      f"Phone number: {self.phone_number}\n"
      f"Email: {self.email}\n"
      f"Country: {self.country}")

  def greet_user(self):
    """Greet the user"""
    print(f"Greetings, {self.first_name} {self.last_name} from {self.country}!")

  def increment_login_attempts(self):
    self.login_attempts += 1

  def reset_login_attempts(self):
    self.login_attempts = 0
```

```python
user = User("Teo", "Spring", 123, "teo@gmail.com", "Argentina")
```

```python
user.increment_login_attempts()
print(user.login_attempts)
```

```python
user.reset_login_attempts()
print(user.login_attempts)
```
